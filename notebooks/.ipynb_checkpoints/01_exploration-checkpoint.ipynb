# Exploration of the PlantVillage Dataset

```python
import zipfile
from pathlib import Path
from src.config import DATA_RAW_DIR, DATA_PROCESSED_DIR

# Path of the raw zip file and the extraction directory
raw_zip = Path(DATA_RAW_DIR) / "plantvillage.zip"
extract_dir = Path(DATA_PROCESSED_DIR)

# Creation of the extraction directory if it doesn't exist
extract_dir.mkdir(parents=True, exist_ok=True)

# Extraction of the zip file
with zipfile.ZipFile(raw_zip, 'r') as z:
    z.extractall(extract_dir)

print(f"Extracted {raw_zip} to {extract_dir}")
```

We have successfully extracted the PlantVillage dataset. Next, we will explore the directory structure and the contents of the dataset.

# Setting Up GPU Memory Growth for TensorFlow

```python
import tensorflow as tf

gpus = tf.config.experimental.list_physical_devices('GPU')
for gpu in gpus:
    tf.config.experimental.set_memory_growth(gpu, True)

print(tf.config.list_physical_devices('GPU'))
```

We have enabled memory growth for the available GPUs. This helps in managing GPU memory usage more efficiently during model training.
